```python
import torch
import numpy as np
import torch.nn.functional as F
from torch import nn
from torch.utils.data import TensorDataset, DataLoader
import torchvision
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import os
import random
import math
from functools import cmp_to_key
import shutil
```

```python
def generate_noise(n:int) -> torch.tensor:
    tmp = torch.randn(n, 3, 64, 64)
    tmp -= torch.min(tmp)
    return tmp / (torch.max(tmp) - torch.min(tmp))

def shuffle(array:[]) -> []:
    n = len(array)
    for i in range(0, n - 1):
        j = random.randint(0, n - 1)
        array[i], array[j] = array[j], array[i]
    return array

def save_image(image:torch.tensor, path:str) -> None:
    transform = torchvision.transforms.ToPILImage()
    image.to("cpu")
    pil_image = transform(image.byte())
    pil_image.save(path, quality=100)
```

```python
def generate_dataset(noise_levels:[float]) -> None:
    clean_img = []
    path = os.path.join(os.getcwd(), "Dataset2")
    path_clean_images = os.listdir(os.path.join(path, "64x64")) 

    def compare(img1:str, img2:str) -> int:
        return int(img1.replace(".jpg", '')) - int(img2.replace(".jpg", ''))

    path_clean_images = sorted(path_clean_images, key = cmp_to_key(compare))

    for rel_img_path in path_clean_images:
        absolute_image_path = os.path.join(path, "64x64", rel_img_path)
        image = torchvision.io.read_image(absolute_image_path)[:3,:,:]
        clean_img.append(image)

    images_path = os.path.join(path, "images")
    for dir in os.listdir(images_path):
        shutil.rmtree(os.path.join(images_path, dir))

    noisy_image_path = os.path.join(images_path, "noised")
    denoised_image_path = os.path.join(images_path, "denoised")
    os.mkdir(os.path.join(noisy_image_path))
    os.mkdir(os.path.join(denoised_image_path))

    counter = 1
    noise_levels = sorted(noise_levels, reverse=True)
    for i in range(0, len(clean_img)):
        image = clean_img[i]
        for j in range(0, len(noise_levels) - 1):
            noise_map = generate_noise(1)[0] * 255.0
            noisy_image = torch.clamp(torch.round(image * (1.0 - noise_levels[j]) + (noise_map * noise_levels[j])), 0, 255).to(torch.uint8).to('cpu')
            denoised_image = torch.clamp(torch.round(image * (1.0 - noise_levels[j + 1]) + (noise_map * noise_levels[j + 1])), 0, 255).to(torch.uint8).to('cpu')

            save_image(noisy_image, os.path.join(noisy_image_path, str(counter) + ".jpg"))
            save_image(denoised_image, os.path.join(denoised_image_path, str(counter) + ".jpg"))
            counter += 1

generate_dataset([0.0, 0.14, 0.27, 0.39, 0.5, 0.6, 0.69, 0.77, 0.84, 0.9, 0.94, 0.97, 1.0])
```

```python
def load_dataset() -> tuple[torch.tensor, torch.tensor]:
    X = []
    Y = []

    path_X = os.path.join(os.getcwd(), "Dataset2", "images", "noised")
    path_Y = os.path.join(os.getcwd(), "Dataset2", "images", "denoised")
    images_path_X = os.listdir(path_X)
    images_path_Y = os.listdir(path_Y)

    for relative_image_path_X in images_path_X:
        absolute_image_path_X = os.path.join(path_X, relative_image_path_X)
        image = torchvision.io.read_image(absolute_image_path_X)[:3,:,:]
        X.append(image)

    for relative_image_path_Y in images_path_Y:
        absolute_image_path_Y = os.path.join(path_Y, relative_image_path_Y)
        image = torchvision.io.read_image(absolute_image_path_Y)[:3,:,:]
        Y.append(image)

    return (X, Y)

X, Y = load_dataset()
```

```python
def plot_dataset_sample(nb_sample:int = 1) -> None:
    for _ in range(0, nb_sample):
        randIndex = random.randint(0, len(X))
        print("Images : " + str(randIndex))
        plt.imshow(X[randIndex].permute(1, 2, 0))
        plt.show()
        plt.imshow(Y[randIndex].permute(1, 2, 0))
        plt.show()

nb_sample_to_plot = 1
plot_dataset_sample(nb_sample_to_plot)
```

```python
class DiffusionModel(nn.Module):
    def __init__(self, in_channels=3, hidden_channels=64, num_blocks=10):
        super(DiffusionModel, self).__init__()

        self.blocks = nn.ModuleList()

        self.blocks.append(self._create_block(in_channels, hidden_channels))

        for _ in range(num_blocks - 2):
            self.blocks.append(self._create_block(hidden_channels, hidden_channels))

        self.blocks.append(self._create_block(hidden_channels, in_channels))

    def _create_block(self, in_channels:int, out_channels:int) -> nn.Sequential:
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU()
        )

    def forward(self, x:torch.tensor) -> torch.Tensor:
        for block in self.blocks:
            x = block(x)
        return x

    def sample(self) -> torch.Tensor:
        input = generate_noise(1)

        output = self(input)
        return output[0]
```

```python
lr = 0.0005
batch_size = 1000
epochs = 12
print_epoch_interval = 1
epoch_to_show_sample = [10, 20, 30, 40, 50]
test_proportion = 0.05
saving_model = True
continue_training = True
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print("Device : " + device)
```

```python
def train(model:DiffusionModel, X_train:[torch.tensor], y_train:[torch.tensor], criterion) -> DiffusionModel:

    X_train = (torch.stack(X_train).float() / 255.0).to(device)
    y_train = (torch.stack(y_train).float() / 255.0).to(device)


    train_dataset = TensorDataset(X_train, y_train)
    train_loader = DataLoader(train_dataset, batch_size = batch_size, shuffle=True)

    model = model.to(device)

    optimizer = torch.optim.Adam(model.parameters(), lr = lr)

    print("Start training Diffusion model")
    for epoch in range(0, epochs):
        print_sample = epoch in epoch_to_show_sample
        for inputs_batch, expected_outputs_batch in train_loader:
            optimizer.zero_grad()
            outputs_batch = model(inputs_batch)
            loss = criterion(outputs_batch, expected_outputs_batch)
            loss.backward()
            optimizer.step()

            if print_sample:
                print_sample = False
                input = (255.0 * inputs_batch[0]).to(torch.uint8).to('cpu')
                output = (255.0 * outputs_batch[0]).to(torch.uint8).to('cpu')
                output_expected = (255.0 * expected_outputs_batch[0]).to(torch.uint8).to('cpu')
                plt.imshow(input.permute(1, 2, 0))
                plt.show()
                plt.imshow(output.permute(1, 2, 0))
                plt.show()
                plt.imshow(output_expected.permute(1, 2, 0))
                plt.show()

        if(epoch == 0 or  (epoch+1) % print_epoch_interval == 0 or epoch + 1 == epochs):
            print(f"Epoch {epoch+1}, Loss: {loss.item()}")

    if saving_model:
        path = "./Models/diffuser.model"
        if os.path.isfile(path):
            os.remove(path)

        torch.save(model.state_dict(), path)
        print(f"Diffusion model Saved!")

    return model

if not continue_training:
    model = DiffusionModel()

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = test_proportion, random_state=145, shuffle=True)
criterion = nn.MSELoss()

model = train(model, X_train, y_train, criterion)
```

```python
def load() -> DiffusionModel:
    model = DiffusionModel()
    model.load_state_dict(torch.load(os.path.join(os.getcwd(), "Models", "diffuser.model")))
    return model

model = load()
```

```python
def test_model(nb_images:int):

    images_inputs = []
    images_outputs = []
    expected_images_outputs = []
    model.to(device)

    for i in range(0, nb_images):

        input = (torch.stack([X_test[i]]).float() / 255.0).to(device)

        output = model(input)

        outputs_images = (255.0 * output).to(torch.uint8).to('cpu')

        images_inputs.append(X_test[i].permute(1, 2, 0))
        images_outputs.append(outputs_images[0].permute(1, 2, 0))
        expected_images_outputs.append(y_test[i].permute(1, 2, 0))

    for i in range(0, len(images_inputs)):
        print(f"Image{i}")
        plt.imshow(images_inputs[i])
        plt.show()
        plt.imshow(images_outputs[i])
        plt.show()
        plt.imshow(expected_images_outputs[i])
        plt.show()

test_model(3)
```

```python
def compute_mean_loss() -> float:
    loss = 0.0

    for k in range(0, len(X_test)):
        inputs =  (torch.stack([X_test[k]]).float() / 255.0).to(device)
        outputs_test = model(inputs)
        loss += criterion(inputs, outputs_test)

    loss /= float(len(X_test))

    return float(loss)

mean_loss = compute_mean_loss()
print(mean_loss)
```

```python
def test() -> None:    

    def show_current_step(step:int, current_img:torch.tensor):
        print("step : " + str(step))
        tmp = (255.0 * current_img).to(torch.uint8).to('cpu').permute(1, 2, 0)
        plt.imshow(tmp)
        plt.show()

    nb_img_test = 1
    model.to(device)
    
    for k in range(0, nb_img_test):

        nb_step = 10

        current_img = generate_noise(1)[0].to(device)
        step = 0
        show_current_step(step, current_img)
        
        for i in range(0, nb_step):

            current_img = model(current_img)
            step += 1

            show_current_step(step, current_img)

test()
```
